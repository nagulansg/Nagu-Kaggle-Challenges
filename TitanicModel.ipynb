## Looking at the number of NaN values and replacing them with the mean or max occurence

```python
df_train.isna().sum()
```

```python
df_train.Age=df_train.Age.fillna(df_train.Age.mean())
#by default value_counts will sort descending
df_train.Embarked=df_train.Embarked.fillna(df_train.Embarked.value_counts().index[0])
```

## Time to do label encoding
###### Survived can be dropped and pushed to y, Pclass (ordinal encoding), sex (One hot),Sibsp & Parch (maybe ordinal maybe one hot),Embarked (one hot encoding)

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
from sklearn.compose import make_column_transformer

X=df_train.drop("Survived",axis=1)
y=df_train["Survived"]

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,
                                               y,random_state=42,stratify=y)



transformer = make_column_transformer(
    (OneHotEncoder(drop="first"),["Sex","Embarked"]),
    remainder='passthrough')

transformed=transformer.fit_transform(X_train)
transformed_train_df=pd.DataFrame(transformed,
                            columns=transformer.get_feature_names_out())
```
