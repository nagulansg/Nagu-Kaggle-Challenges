We do not care about id as it will not help with predictions.
Loading should be a float and product_code should be an object which they are.
Attribute 0-3 should be objects as they are ATTRIBUTES. They can seem numerical but they are more like features or unranked categories. So attribute 0 and 1 are fine but 2 and 3 must be changed to object.
What we can see is that measurement 1 to 17 have correct datatypes, as does failure. Failure is a 0 or 1 value, but we want to predict probabilities of failure therefore we will keep the data type as int64.

```python
#replacing all NaNs with zero
df_train = df_train.fillna(0)
```

```python
X=df_train.drop(columns=["failure","id"],axis=1).copy()
X.head()
```

```python
y=df_train["failure"].copy()
y.head()
```

```python
X_encoded=pd.get_dummies(X,columns=['product_code','attribute_0','attribute_1','attribute_2','attribute_3'])
X_encoded.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_encoded,y,random_state=42,stratify=y)
```

```python
sum(y_train)/len(y_train)
```

```python
sum(y_test)/len(y_test)
```

```python
X_test
```

```python
import xgboost as xgb
#from sklearn.metrics import eval_metrics
#first we create a shell within which we will create a forest of xgb trees
clf_xgb=xgb.XGBClassifier(objective="binary:logistic", missing =0,seed = 42,eval_metric='aucpr',early_stopping_rounds=10)
#then we create the trees
clf_xgb.fit(X_train,y_train,verbose=True,eval_set=[(X_test,y_test)])
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
predictions = clf_xgb.predict(X_test)
accuracy_score(y_test,predictions)
```

```python
from sklearn.model_selection import GridSearchCV
'''
#first bit of tuning
param_grid = {
        'max_depth' : [3,4,5],
        'learning_rate' : [0.1,0.01,0.05],
        'gamma' : [0,0.25,1.0],
        'reg_lambda': [0, 1.0, 10.0],
        'scale_pos_weight' : [1,3,5]
}
'''

#second bit of tuning (comment out first tune after
#you input the optimal ranges into the second param_grid)
#FIRST TUNE SAID GAMMA OF 1, LEARNING RATE OF 0.01, MAX DEPTH OF 3, REG_LAMBDA OF 10 AND SCALE POS WEIGHT OF 1


param_grid = {
        'max_depth' : [1,2,3],
        'learning_rate' : [0.01],
        'gamma' : [1.0,1.5,2],
        'reg_lambda': [10.0,12.0,14.0],
        'scale_pos_weight' : [0.1,0.5,1]
}

#now Fitting 3 folds for each of 81 candidates, totalling 243 fits
#{'gamma': 2, 'learning_rate': 0.01, 'max_depth': 2, 'reg_lambda': 14.0, 'scale_pos_weight': 1}


param_grid = {
        'max_depth' : [2],
        'learning_rate' : [0.01],
        'gamma' : [2,2.5,3],
        'reg_lambda': [14.0,16.0,18.0],
        'scale_pos_weight' : [1,1.5]
}


#josh suggest using subsampling to prevent overfitting
#by using random subsets of the data (90%) and only random subset
#of columns (50%) per tree. 3-fold validation for speed.

optional_params = GridSearchCV(
    estimator=xgb.XGBClassifier(objective='binary:logistic',
                               seed=42,
                               subsample=0.9,
                               colsample_bytree=0.5),
    param_grid=param_grid,
    scoring='roc_auc',
    verbose=2,
    n_jobs=10,
    cv=3
)

optional_params.fit(X_train,y_train)
print(optional_params.best_params_)
```

```python
clf_xgb = xgb.XGBClassifier(seed=42,objective='binary:logistic',gamma=3,learning_rate=0.01,max_depth=2,
                            reg_lambda=18,scale_pos_weight=1,subsample=0.9,colsample_bytree=0.5,
                            early_stopping_rounds=10,eval_metric='aucpr')

clf_xgb.fit(X_train,y_train,verbose=True,eval_set=[(X_test,y_test)])
```
